```python
import sqlite3
import torch
from transformers import pipeline
from deepmultilingualpunctuation import PunctuationModel

torch.cuda.is_available()
```

```python
conn = sqlite3.connect('letters.db')
cursor = conn.cursor()

res = cursor.execute("""
    select * from letters 
    where gestora = 'Encore' and title like '%Comentário%'
""")

res.fetchone()
```

```python
letters = []
cols = [col[0] for col in res.description]

for row in res.fetchall():
    letters.append(dict(zip(cols, row))) 

print(cols)
print(letters[0])
```

```python
conn.close()
```

```python
model = PunctuationModel(model = "kredor/punctuate-all")
result = model.restore_punctuation(letters[0]['content'])

print(letters[0]['content'])
print(result)
```
